```
#Q1 Import the required libraries (sklearn, pandas, numpy). Import datasets. Concatenate the datasets. Display first 5 rows.
import pandas as pd 
import numpy as np
import sklearn
d = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv ")
c = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
t = pd.concat([d,c]).head(5)
t
```

```
#Q2.Remove NaN values from the column ‘Survived’. 

t.Survived.dropna()
```

```
#Q3. Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 
t['Has_Cabin'] = ~t.Cabin.isnull() #'~' symbol is used to store the true and false result in has_cabin column
t.head()
```

```
#Q5 Impute the missing values using .fillna() method
c = t['Age'] = t['Age'].fillna('S')
c
```

```
#Q7 Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.
t.drop(['Cabin','Name','PassengerId','Parch','SibSp'],axis=1)
t.head()
```
